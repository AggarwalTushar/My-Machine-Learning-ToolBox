## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import confusion_matrix
```

## Importing the Dataset

```python
#Reading the data from CSV file
data=pd.read_csv('A_Z Handwritten Data.csv')
print(data.columns.values)
#Generating the features
X=data.iloc[:,1:].values
print(X.shape)
#Generating the output values
Y=data.iloc[:,0].values
```

## Scaling the Dataset

```python
#Converting the values into 0-1 with threshold 128
X=X//128
```

## Splitting the data into train and test

```python
#Splitting the data into 80% training and 20% test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2,random_state=0)
```

## Training the Naive Bayes model on the Training set

```python
#Creating the classifier
classifier = BernoulliNB()
classifier.fit(X_train, Y_train)
```

## Predicting the Test set results

```python
#Predicting the test set values
Y_pred = classifier.predict(X_test)

#Calculating the number of correct values predicted
CC=np.sum(Y_pred==Y_test)
#Printing the accuracy
print("Accuracy: "+str((CC)*100/len(Y_test)))
```

## Generating the Confusion Matrix

```python
#A confusion matrix is a technique for summarizing the performance of a classification algorithm.
cm = confusion_matrix(Y_test, Y_pred)
cm
```
